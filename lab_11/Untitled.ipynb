```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
G= nx.Graph()

G.add_edge("Argentina", "Chile", weight=0.6) 
G.add_edge("Argentina", "Brazil", weight=0.1)
G.add_edge("Brazil", "Paraguay", weight=0.2)
G.add_edge("Brazil", "Columbia", weight=0.2)
G.add_edge("Argentina", "Paraguay", weight=0.3)
G.add_edge("Argentina", "Uruguay", weight=0.2)
G.add_edge("Argentina", "Bolivia", weight=0.2)
G.add_edge("Brazil", "Peru", weight=0.5)
G.add_edge("Brazil", "Uruguay", weight=0.1)
G.add_edge("Brazil", "Venezuela", weight=0.2)
G.add_edge("Brazil", "Bolivia", weight=0.5)
G.add_edge("Chile", "Bolivia", weight=0.2)
G.add_edge("Chile", "Peru", weight=0.1)
G.add_edge("Bolivia", "Peru", weight=0.3)
G.add_edge("Bolivia", "Paraguay", weight=0.2)
G.add_edge("Peru", "Columbis", weight=0.3)
G.add_edge("Peru", "Equador", weight=0.3)
G.add_edge("Columbia", "Equador", weight=0.3)
G.add_edge("Columbia", "Venezuela", weight=0.6)

elarge=[(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] > 0.5]
esmall=[(u, v) for (u, v, d) in G.edges(data=True) if d["weight"]<-0.5] 

pos = nx.spring_layout(G, seed=7)

nx.draw_networkx_nodes(G, pos, node_size=700)

nx.draw_networkx_edges (G, pos, edgelist=elarge, width=6) 
nx.draw_networkx_edges(G, pos, edgelist=esmall, width=6, alpha=0.5, edge_color="b", style="dashed")

nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif") #edge weight Lobels 
                      
edge_labels = nx.get_edge_attributes(G, "weight") 
                      
nx.draw_networkx_edge_labels(G, pos, edge_labels)

ax = plt.gca()
ax.margins(0.08) 
plt.axis("off")
plt.tight_layout() 
plt.show()
```
